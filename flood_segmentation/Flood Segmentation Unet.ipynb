```python
IMG_SIZE = (224, 224) # ukuran target gambar

# mengubah mask menjadi biner (0 atau 1)
def modify_mask(mask):
    mask = np.expand_dims(mask, axis = 2) # menjadikan mask 2D menjadi 3D
    t_mask = np.zeros(mask.shape)
    np.place(t_mask[:, :, 0], mask[:, :, 0] >=0.5, 1) # memberi nilai 1 pada piksel yang nilai intensitasnya >= 0.5
    return t_mask

# membaca gambar dan mask dari path dan melakukan preprocessing
def map_function(img, mask, training):
    img, mask = plt.imread(img.decode())[:, :, :3], plt.imread(mask.decode()) # mengambil 3 channel (RGB) dari gambar
    img = op.resize(img, IMG_SIZE) # resize gambar dan mask ke ukuran IMG_SIZE
    mask = modify_mask(op.resize(mask, IMG_SIZE)) # konversi mask ke bentuk biner
    img = img/255.0 # normalisasi
    if training == True:
        transformed = transform(image=img, mask=mask) # augmentasi
        img = transformed['image']
        mask = transformed['mask']
    
    return img.astype(np.float64), mask.astype(np.float64)
```

```python
BATCH_SIZE = 16 # ukuran batch saat training

# membuat dataset dari data images dan mask
def create_dataset(data, training = True):
    dataset = tf.data.Dataset.from_tensor_slices((data['Image'], data['Mask']))
    dataset = dataset.shuffle(100)
    dataset = dataset.map(lambda img, mask : tf.numpy_function(
                    map_function, [img, mask, training], [tf.float64, tf.float64]),
                    num_parallel_calls = tf.data.experimental.AUTOTUNE).batch(BATCH_SIZE) # membaca file gambar & mask, resize & normalisasi, dan augmentasi
    dataset = dataset.prefetch(buffer_size = tf.data.experimental.AUTOTUNE)
    return dataset
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, test_size = 0.1) # membagi data menjadi train dan test dengan ukuran test 0.1

print(df_train.shape, df_test.shape)
train_dataset = create_dataset(df_train, training = True) # membuat pipeline tf.data.Dataset dari df_train
test_dataset = create_dataset(df_test, training = False) # membuat pipeline tf.data.Dataset dari df_test
```

```python
# menghitung dice coefficient yang bernilai antara 0 dan 1. Semakin dekat ke 1, artinya prediksi dan label sangat mirip.
def dice_coef(y_true, y_pred):
    y_true_f = K.flatten(y_true) # meratakan label menjadi 1D array
    y_pred_f = K.flatten(y_pred) # meratakan prediksi jadi 1D array
    intersection = K.sum(y_true_f * y_pred_f) # jumlah overlap antara label dan prediksi
    return (2. * intersection + 1.0) / (K.sum(y_true_f) + K.sum(y_pred_f) + 1.0) 
 
# menghitung loss dari dice coefficient
def dice_coef_loss(y_true, y_pred):
    return 1-dice_coef(y_true, y_pred)
```

```python
# membuat model U-Net dengan arsitektur backbone EfficientNetB2
model = sm.Unet('efficientnetb2', 
                input_shape = (224,224,3), 
                classes = 1, 
                activation='sigmoid', 
                encoder_weights='imagenet')
```

```python
# compile model
model.compile(
    optimizer = keras.optimizers.Adam(learning_rate = 2e-4),
    loss = keras.losses.BinaryCrossentropy(),
    metrics = [sm.metrics.iou_score],
)
```

```python
# latih model
history = model.fit(train_dataset, validation_data = test_dataset, epochs = 15)
```

```python
# visualisasi nilai loss dan IoU Score
plt.style.use('seaborn')

plt.figure(figsize = (13,5))
plt.subplot(1,2,1)
plt.title('Loss')
plt.plot(history.history['loss'], label = 'Training')
plt.plot(history.history['val_loss'], '--r', marker = 'o', label = 'Validation')
plt.legend()
plt.subplot(1,2,2)
plt.title('IoU Score')
plt.plot(history.history['iou_score'], label = 'Training')
plt.plot(history.history['val_iou_score'], '--r', marker = 'o', label = 'Validation')
plt.legend()
```

```python
color = np.array([161, 205, 255])/255.0 # normalisasi warna

# fungsi untuk menyiapkan mask dari hasil prediksi model
def make_pred_good(pred):
    pred = pred[0][:, :, :]
    pred = modify_mask(pred[:, :, 0])
    pred = np.repeat(pred, 3, 2)
    return pred

# fungsi untuk meletakkan mask di atas gambar asli 
def placeMaskOnImg(img, mask):
    np.place(img[:, :, :], mask[:, :, :] >= 0.5, color)
    return img
```
