```python
# import library yang diperlukan
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, Subset
from torchvision import models, transforms
from skimage.feature import graycomatrix, graycoprops
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
# dataset yang digunakan adalah citra batik 
# dataset memiliki dua kelas yaitu kelas positive untuk citra batik motif kuntul perak dan kelas negative untuk citra batik motif bukan kuntul perak
# dataset yang digunakan berjumlah 200 setiap kelasnya sehingga total kesuluruhan dataset adalah 400 citra

fig, axes = plt.subplots(1, 2, figsize=(8, 4)) 
# menampilkan dataset citra
# citra batik motif kuntul perak/positive
img1 = Image.open('./dataset/positive/1.jpg')
axes[0].imshow(img1)
axes[0].set_title('Gambar Citra Batik Motif Kuntul Perak')
axes[0].axis('off')

# citra batik motif bukan kuntul perak/negative
img2 = Image.open('./dataset/negative/1.jpg')
axes[1].imshow(img2)
axes[1].set_title('Gambar Citra Batik Motif Bukan Kuntul Perak')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
# fungsi untuk membaca image path dan label dari path dataset
def load_image_paths_and_labels(root_dir):
    paths = []
    labels = []
    for label in os.listdir(root_dir):
        folder = os.path.join(root_dir, label)
        for img_file in os.listdir(folder):
            paths.append(os.path.join(folder, img_file))
            labels.append(label)
    return paths, labels
```

```python
# fungsi untuk membaca citra dan convert warna citra
def load_and_convert_image(img_path):
    img_bgr = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB) # mengubah warna bgr menjadi rgb
    img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY) # mengubah warna menjadi grayscale
    return img_rgb, img_gray
```

```python
# fungsi untuk ektraksi fitur menggunakan metode Gray Level Co-occurrence Matrix (GLCM)
def extract_glcm(img_gray):
    glcm = graycomatrix(img_gray, [1], [0], levels=256, symmetric=True, normed=True) # membuat matriks GLCM dari gambar
    features = [
        graycoprops(glcm, prop).ravel()[0] 
        for prop in ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'ASM'] # mengambil fitur statistik dari GLCM
    ]
    return np.array(features, dtype=np.float32)
```

```python
# konversi fitur GLCM (dari NumPy array) menjadi tensor PyTorch 
def extract_texture_features(img_gray):
    return torch.tensor(extract_glcm(img_gray), dtype=torch.float32)
```

```python
# fungsi untuk mengubah label string menjadi numerik
def encode_labels(labels):
    encoder = LabelEncoder()
    encoded = encoder.fit_transform(labels)
    return encoded, encoder.classes_, encoder
```

```python
class BatikDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.paths, raw_labels = load_image_paths_and_labels(root_dir)
        self.labels, self.classes, self.encoder = encode_labels(raw_labels)
        self.transform = transform

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        img_path = self.paths[idx]
        label = self.labels[idx]

        img_rgb, img_gray = load_and_convert_image(img_path)
        img_tensor = self.transform(img_rgb) if self.transform else transforms.ToTensor()(img_rgb)
        texture_feat = extract_texture_features(img_gray)

        return img_tensor, texture_feat, label
```

```python
# model resnet50
class model(nn.Module):
    def __init__(self, num_texture_features=6, num_classes=2):
        super(model, self).__init__()
        self.cnn = models.resnet50(pretrained=True)
        self.cnn.fc = nn.Identity()

        self.classifier = nn.Sequential(
            nn.Linear(2048 + num_texture_features, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, num_classes)
        )

    def forward(self, image, texture_feat):
        cnn_feat = self.cnn(image)
        combined = torch.cat((cnn_feat, texture_feat), dim=1)
        return self.classifier(combined)
```

```python
# preprocessing input gambar sebelum training
transform = transforms.Compose([
    transforms.ToPILImage(), # konversi citra dari NumPy array atau tensor menjadi format PIL
    transforms.Resize((224, 224)), # mengubah ukuran gambar menjadi 224x224 piksel
    transforms.ToTensor(), # mengubah gambar PIL menjadi tensor PyTorch d
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]) # menormalkan gambar dengan mean dan standar deviasi channel RGB dari ImageNet
])
```

```python
# dataset & dataLoader
dataset = BatikDataset("dataset", transform=transform)
train_idx, test_idx = train_test_split(np.arange(len(dataset)), test_size=0.2, stratify=dataset.labels, random_state=42)
train_data = Subset(dataset, train_idx)
test_data = Subset(dataset, test_idx)
train_loader = DataLoader(train_data, batch_size=16, shuffle=True)
test_loader = DataLoader(test_data, batch_size=16, shuffle=False)
```

```python
# inisialisasi model & optimizer
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
```

```python
# training
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    total_loss, correct, total = 0, 0, 0

    for images, texture_feats, labels in train_loader:
        images, texture_feats, labels = images.to(device), texture_feats.to(device), labels.to(device)

        outputs = model(images, texture_feats)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        avg_loss = total_loss / len(train_loader)
        _, predicted = torch.max(outputs, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

    acc = correct / total * 100
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}, Accuracy: {acc:.2f}%")
```

```python
# Evaluasi
model.eval()
correct, total = 0, 0
with torch.no_grad():
    for images, texture_feats, labels in test_loader:
        images, texture_feats, labels = images.to(device), texture_feats.to(device), labels.to(device)
        outputs = model(images, texture_feats)
        _, predicted = torch.max(outputs, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

print(f"Test Accuracy: {correct / total * 100:.2f}%")
```

```python
def predict_and_show_image(model, image_path, transform, encoder):
    model.eval()
    
    # Baca dan konversi gambar
    img_bgr = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

    # Transformasi dan ekstraksi fitur
    img_tensor = transform(img_rgb).unsqueeze(0).to(device)
    texture_feat = extract_texture_features(img_gray).unsqueeze(0).to(device)

    # Prediksi
    with torch.no_grad():
        output = model(img_tensor, texture_feat)
        _, predicted = torch.max(output, 1)

    predicted_label = encoder.inverse_transform(predicted.cpu().numpy())[0]

    # Tampilkan gambar + hasil
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.title(f'Prediksi: {predicted_label}')
    plt.show()

    return predicted_label
```

```python
# tes model menggunakan citra motif batik bukan kuntul perak
image_path = "image_negative.jpg"
hasil = predict_and_show_image(model, image_path, transform, dataset.encoder)
print(f"Hasil prediksi: {hasil}")
```

```python
# tes model menggunakan citra motif batik kuntul perak
image_path = "image_positive.jpg"
hasil = predict_and_show_image(model, image_path, transform, dataset.encoder)
print(f"Hasil prediksi: {hasil}")
```
